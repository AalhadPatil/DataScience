```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('Universities.csv')
```

```python
df1
```

```python
plt.hist(df1['SAT'])
plt.xlabel('SAT')
plt.ylabel('Frequency')
plt.title('Distribution of SAT Scores')
plt.show()
```

```python
sns.histplot(df1['SAT'])
plt.xlabel('SAT')
plt.ylabel('Frequency')
plt.title('Distribution of SAT Scores (Seaborn)')
plt.show()
```

```python
plt.plot(df1['SAT']),df1['Accept']
```

```python
plt.plot(df1['SAT'],df1['Accept'],'bo')
```

```python
plt.xlabel('SAT')
plt.ylabel('Accept')
plt.title('SAT vs Accept')
plt.show()
```

```python
plt.scatter(df1['SAT'], df1['Top10'])
plt.xlabel('SAT Score')
plt.ylabel('Top 10% High School Class')
plt.title('SAT Score vs. Top 10% High School Class')
plt.show()
```

```python
sns.pairplot(df1)
```

```python
sns.pairplot(df1.iloc[:,1:3])
plt.boxplot(df1['SAT'])
plt.show()
```
